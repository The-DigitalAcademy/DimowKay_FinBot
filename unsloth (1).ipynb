```python
import pandas as pd


# Assuming 'train_data.csv' is your data file
train_data = pd.read_csv('./data/train_data.csv')  # This line creates a DataFrame

# Now you can use the rename method
train_data = train_data.rename(columns={'question': 'query', 'answer': 'response'})
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the model and tokenizer
model_name = "unsloth/Llama-3.2-1B-bnb-4bi"  # Replace with the model you want to use
```

```python

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def tokenize_function(examples):
    return tokenizer(examples['query'], padding="max_length", truncation=True)

# Apply tokenization to the dataset
train_dataset = train_dataset.map(tokenize_function, batched=True)
```

```python
from transformers import Trainer, TrainingArguments

# Define training arguments
training_args = TrainingArguments(
    output_dir="./finetuned_model",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir='./logs',
)

# Initialize the trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)

# Start the training process
```

```python
trainer.train()
```
